# Mie Scattering


Giving that
\begin{align}
    x &= k_1R\\
    Mx &= \frac{n_2}{n_1}k_1R = k_2R.
\end{align}
$R$ being the sphere radius, $n_1$ the extern refraction index and $n_2$ inside the sphere.
\begin{align}
    \Psi_n(\rho) &= \rho j_n(\rho)\\
    \zeta(\rho) &= \rho h_n^{(1)}(\rho),
\end{align}
where
\begin{equation}
    j_n(r) = \sqrt{\frac{\pi}{2r}}J_{n+1/2}(r),
\end{equation}

\begin{equation}
    y_n(r) = \sqrt{\frac{\pi}{2r}}Y_{n+1/2}(r),
\end{equation}

\begin{align}
    h_n^{(1)}(r) &= j_n(r)+iy_n(r)\\
    h_n^{(2)}(r) &= j_n(r)-iy_n(r).
\end{align}


Leading us to the coeffcients
\begin{align}
    a_{nm} &= \frac{M\Psi_n(Mx)\Psi_n^{\prime}(x)-\Psi_n^{\prime}(Mx)\Psi_n(x)}{\Psi_n^{\prime}(Mx)\zeta_n(x)-M\Psi_n(Mx)\zeta_n^{\prime}(x)}G_{nm}^{MT}\\
    b_{nm} &=\frac{M\Psi_n^{\prime}(Mx)\Psi_n(x)-\Psi_n(Mx)\Psi_n^{\prime}(x)}{\Psi_n(Mx)\zeta_n^{\prime}(x)-M\Psi_n^{\prime}(Mx)\zeta_n(x)}G_{nm}^{ET}\\
    c_{nm} &= \frac{\Psi_n(x)\zeta_n^{\prime}(x)-\Psi_n^{\prime}(x)\zeta_n(x)}{\Psi_n^{\prime}(Mx)\zeta_n(x)-M\Psi_n(Mx)\zeta_n^{\prime}(x)}G_{nm}^{ET}\\
    d_{nm} &= \frac{\Psi_n^{\prime}(x)\zeta_n(x)-\Psi_n(x)\zeta_n^{\prime}(x)}{\Psi_n(Mx)\zeta_n^{\prime}(x)-M\Psi_n^{\prime}(Mx)\zeta_n(x)}G_{nm}^{MT},
\end{align}


Now that we know how the expected solution looks, we can introduce a vector function called the spherical harmonic vector, where by defining $\vec{L} = i\vec{r}\times \vec{\nabla} $ is

\begin{equation}
    \vec{X}_{nm}(\theta,\phi) = \frac{1}{\sqrt{n(n+1)}}\vec{L}Y_{nm}(\theta,\phi).
\end{equation}

And

\begin{align}
    G_{nm}^{TE}j_n(kr) &= \frac{-k}{E_0\sqrt{n(n+1)}}\int Y_{nm}^*(\theta,\phi)\vec{r}\cdot\vec{E}d\Omega\\
    G_{nm}^{TM}j_n(kr) &= \frac{k}{H_0\sqrt{n(n+1)}}\int Y_{nm}^*(\theta,\phi)\vec{r}\cdot\vec{H}d\Omega.
\end{align}

Representing the fields using the spherical solution of the wave equation, we can express the incident field using constants $G_{nm}^{TE} = A_{nm}$ and $G_{nm}^{TM}=B_{nm}$ for the transversal electric and transversal magnetic components, respectively.

\begin{align}
    \vec{E}_{\text{inc}} &= E_0 \sum_{n,m}( \frac{i}{k_1}G_{nm}^{TM}\nabla \times (j_n(k_1r)\vec{X}_{nm}(\theta,\phi) 
    \\&  + G_{nm}^{TE}j_n(k_1r)\vec{X}_{nm}(\theta,\phi) )\\
    \vec{H}_{\text{inc}} &= \frac{E_0}{Z_1}\sum_{n,m} (G_{nm}^{TM}j_n(k_1r)\vec{X}_{nm}(\theta,\phi) \\
    & -\frac{i}{k_1} G_{nm}^{TE}\nabla \times j_n(k_1r)\vec{X}_{nm}(\theta,\phi) ).
\end{align}

Similarly, we express the field inside the sphere using the refractive index within this space, resulting in constants $c_{nm}$ and $d_{nm}$.

\begin{align}
    \vec{E}_{\text{int}} &= E_0 \sum_{n,m}( \frac{i}{k_2}c_{nm}\nabla \times (j_n(k_2 r)\vec{X}_{nm}(\theta,\phi)) \\
    &  + d_{nm}j_n(k_2r)\vec{X}_{nm}(\theta,\phi) \\
    \vec{H}_{\text{int}} &= \frac{E_0}{Z_2}\sum_{n,m} (c_{nm}j_n(k_2r)\vec{X}_{nm}(\theta,\phi) \\
    &  -\frac{i}{k_2} d_{nm}\nabla \times j_n(k_2r)\vec{X}_{nm}(\theta,\phi) ).
\end{align}

The scattered field exists outside the sphere and propagates through high values of $r$. In this limit, the spherical Hankel function is used, specifically $h_n^{(1)}$, which describes functions that continue growing. The scattered fields are expressed using constants $a_{nm}$ and $b_{nm}$.

\begin{align}
    \vec{E}_{\text{sca}} &= E_0 \sum_{n,m}( \frac{i}{k_1}a_{nm}\nabla \times (h_n^{(1)}(k_1 r)\vec{X}_{nm}(\theta,\phi) \\ 
    &  + b_{nm}h_n^{(1)}(k_1r)\vec{X}_{nm}(\theta,\phi) )\\
    \vec{H}_{\text{sca}} &= \frac{E_0}{Z_1}\sum_{n,m} (a_{nm}h_n^{(1)}(k_1r)\vec{X}_{nm}(\theta,\phi) \\ &  -\frac{i}{k_1} b_{nm}\nabla \times h_n^{(1)}(k_1r)\vec{X}_{nm}(\theta,\phi) ).
\end{align}

Rearanging the Scattered Field we can obtain 

\begin{equation}
    \vec{E_{scat}} = E_0 \sum_{n,m}...
\end{equation}

Substituting the Mie COefficients in the calculations

\begin{align}
    a_{nm} &= \frac{M\Psi_n(Mx)\Psi_n^{\prime}(x)-\Psi_n^{\prime}(Mx)\Psi_n(x)}{\Psi_n^{\prime}(Mx)\zeta_n(x)-M\Psi_n(Mx)\zeta_n^{\prime}(x)}G_{nm}^{MT}\\
    b_{nm} &=\frac{M\Psi_n^{\prime}(Mx)\Psi_n(x)-\Psi_n(Mx)\Psi_n^{\prime}(x)}{\Psi_n(Mx)\zeta_n^{\prime}(x)-M\Psi_n^{\prime}(Mx)\zeta_n(x)}G_{nm}^{ET} ,   
\end{align}
where $G_{n\pm 1}^{ET} = -i \sqrt{\pi (2n +1)} i^{n+1}$, 

$G_{n,+1}^{TM} =-\sqrt{\pi(2n +1)}i^{n+1} $ and 

 $G_{n,-1}^{TM} =\sqrt{\pi(2n +1)}i^{n+1} $

```python
import numpy as np
from scipy.special import spherical_jn, spherical_yn , hankel1,lpmv
```

```python
def MieCoef(x, M, n):
    # Computing spherical Bessel functions for n and M*n
    j_n_x = spherical_jn(n, x)
    j_n_Mx = spherical_jn(n, M*x)

    # Computing Riccati-Bessel functions
    psi_n_x = x * j_n_x
    zeta_n_x = x * np.sqrt(np.pi/(2*x))*hankel1(n+1/2,x)

    psi_n_Mx = M * x * j_n_Mx
    zeta_n_Mx = M * x * np.sqrt(np.pi/(2*M*x))*hankel1(n+1/2,M*x)

    # Computing Riccati-Bessel function derivatives using numdifftools
    #psi_n_x_prime = nd.Derivative(lambda x: x * spherical_jn(n, x), n=1, order=1)(x)
    psi_n_x_prime = 1/2* (  x* (spherical_jn(n-1,x)-spherical_jn(n+1,x)) + spherical_jn(n,x) ) 
    
    zeta_n_x_prime = 1/2* np.sqrt(np.pi/(2*x))*(  x* (hankel1(n+1/2-1,x)-hankel1(n+1/2+1,x)) + hankel1(n+1/2,x) )

    #psi_n_Mx_prime = nd.Derivative(lambda x: M * x * spherical_jn(n, M * x), n=1, order=1)(x)
    psi_n_Mx_prime = 1/2* (  M*x* (spherical_jn(n-1,M*x)-spherical_jn(n+1,M*x)) + spherical_jn(n,M*x) ) 
    
    zeta_n_Mx_prime= 1/2* np.sqrt(np.pi/(2*M*x))*(  x* (hankel1(n+1/2-1,M*x)-hankel1(n+1/2+1,M*x)) + hankel1(n+1/2,M*x) )

    # Calculating coefficients a_n, b_n, c_n, d_n
    a_n = (M * psi_n_Mx * psi_n_x_prime - psi_n_Mx_prime * psi_n_x) / (psi_n_Mx_prime * zeta_n_x - M * psi_n_Mx * zeta_n_x_prime)
    b_n = (M * psi_n_Mx_prime * psi_n_x - psi_n_Mx * psi_n_x_prime) / (psi_n_Mx * zeta_n_x_prime - M * psi_n_Mx_prime * zeta_n_x)
    
    #c_n = (psi_n_x * zeta_n_x_prime - psi_n_x_prime * zeta_n_x) / (psi_n_Mx_prime * zeta_n_x - M * psi_n_Mx * zeta_n_x_prime)
    #d_n = (psi_n_x_prime * zeta_n_x - psi_n_x * zeta_n_x_prime) / (psi_n_Mx * zeta_n_x_prime - M * psi_n_Mx_prime * zeta_n_x)

    return a_n, b_n

# Test the function with scalar inputs
wavelength = 1
radius = 5
k = 2 * np.pi / wavelength
x_value = k * radius
print(MieCoef(x_value, 155/133, 30))
```

Definindo

\begin{equation}
    Y_{nm} = \sqrt{ \frac{2n+1}{4\pi}\frac{(n-m)!}{(n+m)!}}P_n^m(cos\theta)e^{im\phi}
\end{equation}

e como python não tem hankel esférico, podemos construir usando o hankel
\begin{equation}
    h_n^{(1)}(z) = \sqrt{ \frac{\pi}{2z}}H_{n+1/2}^{(1)}(z)
\end{equation}


```python
def E_SCATnm(n, a_n, b_n, k, r, theta, phi):
    kr = k*r

    hn = np.sqrt(np.pi/(2*kr))*hankel1(n+1/2,kr)
    hn1 = np.sqrt(np.pi/(2*kr))*hankel1(n+1/2+1,kr)
    Pn1 = lpmv(1,n+1,np.cos(theta))
    Pn = lpmv(1,n,np.cos(theta))

    Gn = 1j**(n) 
    prefac = -Gn/(n*(n+1)) * (2*n+1)*1j

    term_r = -a_n/kr * hn  *Pn*n*(n+1)*np.cos(phi)

    term_the = (-a_n/kr* ( (n+1)*hn - kr*hn1 )* (n*Pn1-(n+1)*np.cos(theta)*Pn)/np.sin(theta) -1j* b_n * hn *Pn/np.sin(theta))*np.cos(phi)
    
    term_phi = (a_n/kr* ( (n+1)*hn - kr*hn1 )* Pn/np.sin(theta) +1j* b_n * hn *(n*Pn1-(n+1)*np.cos(theta)*Pn)/np.sin(theta))* np.sin(phi)
        
    return prefac*np.array([ term_r, term_the ,term_phi ])
    
```

```python
def MieCoefCond2(x, n, k1):
    jn = spherical_jn(n, x)
    jn1 = spherical_jn(n+1,x)
    jn_1 = spherical_jn(n-1,x)
    
    hankeln = np.sqrt(np.pi/(2*x))*hankel1(n+1/2,x)
    hankeln1 = np.sqrt(np.pi/(2*x))*hankel1(n+1+1/2,x)
    hankeln_1 = np.sqrt(np.pi/(2*x))*hankel1(n-1+1/2,x)
    
    a = -(jn + x *1/2* (  x* (jn_1-jn1) + jn ))/(hankeln + x  *1/2* ( x*(hankeln_1-hankeln1) + hankeln))

    b = - jn/hankeln  

    
    return a, b

# Example usage
x = 1.0
n = 1
GTM = 2.0
GTE = 3.0
k1 = 0.5


a, b = MieCoefCond2(x, n, k1)
print("a:", a)
print("b:", b)
```

# Plot in Cartesian coordinates


knowing

\begin{align}
\hat{r} &= \cos\phi\sin\theta\hat{x}+\sin\phi\sin\theta\hat{y}+\cos\theta\hat{z} \\
\hat{\phi} &= -\sin\phi\hat{x} + \cos\phi\hat{y} \\
\hat{\theta} &= \cos\phi\cos\theta\hat{x} + \sin\phi\cos\theta\hat{y} -\sin\theta\hat{z}
\end{align}

```python
def E1_SCATnm(n, a_n, b_n, k, x, y, z):
    r = np.sqrt(x**2 + y**2 + z**2)
    phi = np.sign(y) * np.arccos(x/np.sqrt(x**2+y**2))
    theta = np.arccos(z / r)

    kr = k * r
    hn = np.sqrt(np.pi / (2 * kr)) * hankel1(n + 1 / 2, kr)
    hn1 = np.sqrt(np.pi / (2 * kr)) * hankel1(n + 1 / 2 + 1, kr)
    Pn1 = lpmv(1, n + 1, np.cos(theta))
    Pn = lpmv(1, n, np.cos(theta))

    dP = (n * Pn1 - (n + 1) * np.cos(theta) * Pn) / np.sin(theta)
    P = Pn / np.sin(theta)

    # Replace zero values of dP and P at theta=0 with -1/2*n*(n+1)

    zero_theta_indices = (np.where(abs(theta) < 10**(-10))[0])
    dP[zero_theta_indices] = -0.5 * n * (n + 1)
    P[zero_theta_indices] = -0.5 * n * (n + 1)

    zero_theta_indices = (np.isnan(phi))
    phi[zero_theta_indices] = 0
    
    Gn = 1j**n
    prefac = -Gn / (n * (n + 1)) * (2 * n + 1) * 1j

    term_r = -a_n / kr * hn * Pn * n * (n + 1) * np.cos(phi)
    term_the = (-a_n / kr * ((n + 1) * hn - kr * hn1) * dP - 1j * b_n * hn * P) * np.cos(phi)
    term_phi = (a_n / kr * ((n + 1) * hn - kr * hn1) * P + 1j * b_n * hn * dP) * np.sin(phi)

    return prefac * (np.array([term_r * np.cos(phi) * np.sin(theta), term_r * np.sin(theta) * np.sin(phi), term_r * np.cos(theta)]) +
                     np.array([term_the * np.cos(phi) * np.cos(theta), term_the * np.cos(theta) * np.sin(phi), -term_the * np.sin(theta)]) +
                     np.array([-term_phi * np.sin(phi), term_phi * np.cos(phi), np.zeros_like(r)]))
```
